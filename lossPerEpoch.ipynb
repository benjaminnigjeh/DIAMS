# Import libraries and dependencies

```python
import numpy as np
from sklearn.model_selection import LeaveOneOut
import pandas as pd
import tensorflow as tf
from google.colab import files
from keras.layers import Dropout
```

# Upload training dataset

```python
uoloaded = files.upload_file('trainDataset')
```

# Training dataset wrangling

```python
file_path = '/content/trainDataset'
df = pd.read_csv(file_path)
Y = df.pop("target")
X = np.array(df)
y = np.array(Y)
```

# Leave one out cross validaion set-up

```python
loo = LeaveOneOut()
loo.get_n_splits(X)
```

# Training model with increasing EPOCH
the training loss and validation loss were averaged after all instances of the leave one out cross validation

```python
EPOCH = 10
validation_loss = []
train_loss = []

for train_index, validation_index in loo.split(X):
    X_train, X_validation = X[train_index], X[validation_index]
    y_train, y_validation = y[train_index], y[validation_index]
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(150, activation=tf.nn.relu))
    model.add(Dropout(0.2))
    model.add(tf.keras.layers.Dense(200, activation=tf.nn.relu))
    model.add(tf.keras.layers.Dense(2, activation=tf.nn.softmax))
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics='accuracy')
    model.fit(X_train, y_train, epochs=EPOCH)
    validation_loss_t, ___ = model.evaluate(X_validation, y_validation)
    train_loss_t, ___ = model.evaluate(X_train, y_train)
    validation_loss.append(validation_loss_t)
    train_loss.append(train_loss_t)

print('the average training loss after', EPOCH, 'epoch:', np.average(train_loss))
print('the average validation loss after', EPOCH, 'epoch:',np.average(validation_loss))
```
